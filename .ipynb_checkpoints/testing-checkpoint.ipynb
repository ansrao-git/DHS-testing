```python
import pandas as pd

data = pd.read_csv('https://raw.githubusercontent.com/ansrao-git/DHS-testing/main/childrecode.csv')

weight_age_sd = 'hw8'
height_age_sd = 'hw5'
weight_height_sd = 'hw11'

#new WHO values
weight_age_sd_new = 'hw71'
height_age_sd_new = 'hw70'
weight_height_sd_new = 'hw72'

# create stunting column
data['stunting'] = data[height_age_sd_new] < -200
#data['severe_stunting'] = data[height_age_sd_new] < -300

# create wasting column
data['wasting'] = data[weight_height_sd_new] < -200
#data['severe_wasting'] = data[weight_height_sd_new] < -300

# create underweight column
data['underweight'] = data[weight_age_sd_new] < -200
#data['severe_underweight'] = data[weight_age_sd_new] < -300

data = data[data['hw2'] <= 9997] # Removing missing
data = data[data['hw3'] <= 9997] # Removing missing

#Divide by 10 to remove implied decimal place
data['hw3'] = data['hw3'] / 10
data['hw2'] = data['hw2'] / 10

df_new = pd.DataFrame({
    'case_id': data['caseid'],
    'weight(kg)': data['hw2'],
    'age': data['b8'],
    'height(cm)': data['hw3'],
    'stunting': data['stunting'],
    'wasting': data['wasting'] ,
    'underweight': data['underweight']
})


df_new.head()
```

```python
target_vars = ['stunting', 'wasting', 'underweight','target']
predictor_vars = data.columns.difference(['hw70', 'hw71', 'hw72', 'underweight', 'stunting', 'wasting','target']).tolist()

data['target'] = (data['hw70'] < -200) & (data['hw71'] < -200) & (data['hw72'] < -200)

# Compute the correlation matrix
corr_matrix = data.corr()

# Extract the correlations between the predictor variables and the target variables
target_corrs = corr_matrix.loc[:, target_vars]

# Print the correlations
corr_target = abs(corr_matrix[target_vars]).mean(axis=1)
corr_target = corr_target.drop(['hw70', 'hw71', 'hw72','hw4','hw7','hw10', 'underweight', 'stunting', 'wasting'])
corr_target_sorted = corr_target.nlargest(20)



#print(target_corrs)
```

```python
#Correlation between predictors and mean of sd of the targets
print(corr_target_sorted)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
# select only cases with non-missing predictor and target variables
#remove nulls
data_new = data[predictor_vars + target_vars]

data_new = data_new.fillna(0)

# Convert the 'target' column to integer type
#data_new[['target','stunting', 'wasting', 'underweight']] = data_new[['target','stunting', 'wasting', 'underweight']].astype(int)
# convert columns to numeric
data_new[predictor_vars + target_vars] = data_new[predictor_vars + target_vars].apply(pd.to_numeric, errors='ignore')

data_new[:] = data_new[:].replace('[^0-9\.]+', '', regex=True).astype(float)


X = data_new[predictor_vars]
y = data_new['target']

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Initialize and fit the logistic regression model

logreg = LogisticRegression(solver='saga', max_iter=1000)
logreg.fit(X_train, y_train)

# Predict the target variable using the trained model
y_pred = logreg.predict(X_test)
accuracy = logreg.score(X_test, y_test)

print("Accuracy = "+str(accuracy))

```
